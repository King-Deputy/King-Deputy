### Functions
Functions organize your code and avoid redundancy. In this lesson, you will learn how to organize your code with functions. A function is a block of code designed to perform a specific task. As you'll see, functions will let you do roughly the same calculation multiple times without duplicating any code.

#### Intro to functions: a simple example
We begin with a simple example of a function. The add_two() function below accepts any number, adds them up, and then returns the result

```python
# writing basic functions
def add_two(a, b):
    """
    adds a and b
    and returns the sum
    """
    summed = a + b
    return summed
```

```python
# using the function
add_two(3, 4)
```

```python
# using the function to combine strings
add_two('Peace', ' Ejimbe')
```

#### print vs return
A print statement and a return statement serve different purposes in a function:
A print statement is used to display information on the console or terminal.
It is used for outputting data to the screen, typically for debugging or providing information to the user.
The data or expressions inside a print statement are shown on the screen but do not affect the flow of the program or the value returned by the function.
print is often used to inspect the intermediate values of variables or communicate results to the user.

```python
# using a print statement within a function
def greet2(name):
    print(f'hello, {name}')
```

```python
# whenever the funtion is called, it prints a result and performs no other function
msg2 = greet2('Tobi')
```

```python
# the variable: msg2 holds no value
print(msg2)
```

return Statement:

A return statement is used to specify the value that a function should return to the caller.
When a function contains a return statement, it immediately ends the execution of the function and passes the specified value back to the calling code.
The returned value can be used in further computations or assigned to variables.
Functions often use return to provide the result of a computation.

```python
def no_args():
    return 'hello'
```

```python
no_args()
```

```python
# functions can take as many arguments as possible
def volume(length, height, breadth):
    return length * height * breadth
```

```python
volume(6,7,8)
```

```python
# functions alo return as many values as possible
def volume2(length, height, breadth):
    return length, height, breadth
```

```python
volume2(7,8,9)
```

```python
# the print function has the sep argument that can be changed
print('Kachi','Ibezim', sep='---')
```

```python
# functions can take lists (or any other data structure) as arguments and manipulate them
def add_lists(lst1, lst2):
    list3 = [i+j for i,j in zip(lst1, lst2)]
    return list3
```

```python
list1 = [1, 2, 3, 4, 6, 8, 9, 11]
list2 = [2, 4, 6, 11, 5, 67, 98, 2]
```

```python
add_lists(list1, list2)
```
