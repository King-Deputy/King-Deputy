QUESTION 1

1. Gender:
- What is the distribution of genders among the students?
- Are there any gender differences in the prevalence of mental health issues?

```python
# Distribution of genders amount the students
mental_health.groupby('Choose your gender')['Choose your gender'].count()
```

```python
# Percentage Distribution of the genders
mental_health['Gender'].value_counts(normalize=True) * 100
```

```python
#Number of students experiencing depression in each gender
mental_health.groupby(['Gender', 'Experiencing Depression?'])['Experiencing Depression?'].count()
```

```python
# Percentage Distribution of the genders expereincing depression
mental_health.groupby('Gender')['Experiencing Depression?'].value_counts(normalize=True) * 100
```

```python
#Number of students experiencing Anxiety in each gender
mental_health.groupby(['Gender', 'Experiencing Anxiety?'])['Experiencing Anxiety?'].count()
```

```python
# Percentage Distribution of the genders expereincing anxiety
mental_health.groupby('Gender')['Experiencing Anxiety?'].value_counts(normalize=True) * 100
```

```python
#Number of students experiencing Panic attacks in each gender
mental_health.groupby(['Gender', 'Experiencing Panic Attacks?'])['Experiencing Panic Attacks?'].count()
```

```python
# Percentage Distribution of the genders expereincing panic attacks
mental_health.groupby('Gender')['Experiencing Panic Attacks?'].value_counts(normalize=True) * 100
```
